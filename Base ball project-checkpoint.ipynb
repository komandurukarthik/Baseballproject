##### IDENTIFYING OUTLIERS

```python
plt.figure(figsize=(12,15),facecolor='white')
plotnumber=1
for column in data:
    if plotnumber<=17:
        ax=plt.subplot(4,5,plotnumber)
        sns.boxplot(data[column])
        plt.xlabel(column,fontsize=20)
        plotnumber+=1
plt.tight_layout()
```

So there are outliers in  Runs,ERA,SHO,SV and E columns which need to be removed.

Let us construct a Co-relation matrix to identify the relation between all the columns with our target variable.

```python
corr=data.corr()
plt.figure(figsize=(20,10))
sns.heatmap(corr,annot=True)
```

The columns ERA and ER and RA has highly co-relation with the wins where we can remove any of them let us remove ERA  as it is having high co-relation.

```python
df=data.drop(columns=['ERA'])
df
```

Let us remove the outliers and Skewness 

###### Outlier removal

first let us find the skewness on all the columns

```python
df.skew()
```

The columns Runs,CG,SHO,SV,E are highly skewed let us remove the skew ness using Log transform

First let us remove the outliers for this iam using Zscore.

```python
from scipy.stats import zscore
import numpy as np
z=np.abs(zscore(df))
threshold=3
np.where(z>3)
```

```python
df_new=df[(z<3).all(axis=1)]
df_new
```

checking for data loss

```python
df_new.shape
```

```python
data_loss=((30-29)/30)*100
```

```python
data_loss
```

only three percentage of data is lost so we can proceed

##### seperating the columns in to Features and Target variable

```python
x=df_new.drop(columns=['W'])
y=df_new['W']
```

###### Scaling the data using Min Max scaler

```python
from sklearn.preprocessing import MinMaxScaler
mms=MinMaxScaler()
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split
x_scaled=mms.fit_transform(x)
```

```python
x_scaled
```

```python
x_train,x_test,y_train,y_test=train_test_split(x_scaled,y,test_size=0.50,random_state=21)
```

Let us initiate our model and train it

```python
regression=LinearRegression()
```

```python
regression.fit(x_train,y_train)
```

Let us find the R squared value for the Trained model

```python
regression.score(x_test,y_test)
```

```python
y_pred=regression.predict(x_test)
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)
```

our model gives 43% accuracy let us use regularization technqiues

```python
from sklearn.linear_model import Ridge,Lasso,RidgeCV,LassoCV
lasscv=LassoCV(alphas=None,max_iter=1000,normalize=True)
lasscv.fit(x_train,y_train)
```

```python
alpha=lasscv.alpha_
alpha
```

```python
lasso_reg=Lasso(alpha)
lasso_reg.fit(x_train,y_train)
```

```python
lasso_reg.score(x_test,y_test)
```

Lasso regression gives me 55% of the score

Using RidgeCV

```python
ridgecv=RidgeCV(alphas=np.arange(0.001,0.1,0.01),normalize=True)
ridgecv.fit(x_train,y_train)
```

```python
ridgecv.alpha_
```

```python
ridge_model=Ridge(alpha=ridgecv.alpha_)
ridge_model.fit(x_train,y_train)
```

```python
ridge_model.score(x_test,y_test)
```

Ridge cv gives us 74% Of accuracy

##### Applying ensemble Techniques

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestRegressor
parameters={'criterion':['mse','mae'],'max_features':["auto","sqrt","log2"]}
rf=RandomForestRegressor()
clf=GridSearchCV(rf,parameters)
clf.fit(x_train,y_train)
print(clf.best_params_)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
import statsmodels.api as sm
rf=RandomForestRegressor(criterion="mse",max_features="sqrt")
rf.fit(x_train,y_train)
rf.score(x_train,y_train)
pred_decision=rf.predict(x_test)
rfs=r2_score(y_test,pred_decision)
print('R2 SCORE:',rfs*100)
rfscore=cross_val_score(rf,x,y,cv=5)
rfc=rfscore.mean()
print("Cross val score:",rfc*100)
```

Our R2 score gives 41% of score so it is the best model|
